```python
# Edge Detection - Bir nesnenin nerde başlayıp bittiğini gösteriyor
```

```python
import cv2
import numpy as np
image = cv2.imread('input.jpg',0)

height,width = image.shape[:2]

sobel_x = cv2.Sobel(image,cv2.CV_64F,0,1,ksize = 5)
sobel_y = cv2.Sobel(image,cv2.CV_64F,1,0,ksize = 5)

cv2.imshow('Rotated Image',image)
cv2.waitKey(0)
cv2.imshow('Sobel X',sobel_x)
cv2.waitKey(0)
cv2.imshow('Sobel Y',sobel_y)
cv2.waitKey(0)

sobel_OR = cv2.bitwise_or(sobel_x,sobel_y)
cv2.imshow('sobel_OR',sobel_OR)
cv2.waitKey(0)

laplacian = cv2.Laplacian(image,cv2.CV_64F)
cv2.imshow('Laplacian',laplacian)
cv2.waitKey(0)

canny = cv2.Canny(image,50,120) #resimdeki
cv2.imshow('Canny',canny)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
def sketch(image):
    img_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    img_gray_blur = cv2.GaussianBlur(img_gray,(5,5),0) #bulanıklık veriyor
    canny_edges = cv2.Canny(img_gray_blur,10,70) # canny ile vücudun bittiği yerini buluyor
    ret,mask = cv2.threshold(canny_edges,70,255,cv2.THRESH_BINARY_INV) #ufak bir değişikliği görmezden geliyor
    return mask

cap = cv2.VideoCapture(0) # 0 yazınca ön kamerayı alıyor
while True:
    ret,frame = cap.read()
    cv2.imshow('Our live Sketcher',sketch(frame))
    if cv2.waitKey(1) == 13: #programdan çıkmak için enter a basıyoruz
        break
        
cap.release()
cv2.destroyAllWindows()
print('done')
```

```python
# Where is Waldo - topluluk içinde insan bulma
```

```python
import cv2
import numpy as np

image=cv2.imread("WaldoBeach.jpg")

cv2.imshow("Where is Waldo", image)
cv2.waitKey(0)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

template = cv2.imread("Waldo.jpg", 0)
cv2.imshow("Waldo", template)
cv2.waitKey(0)

result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(result)

top_left=maxLoc
bottom_right = (top_left[0] + 50, top_left[1]+50) #bulduğu noktadan cerceve içini alması için 
cv2.rectangle (image, top_left, bottom_right, (0,0,255),5)

cv2.imshow("Where is Waldo", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

#karmaşık bir fotoğraf ve bulmak istediğimiz kişi/ nesne fotoğraf veriyoruz ve buluyor.
```

```python
import cv2
cap = cv2.VideoCapture("airplanes.mp4")
while True:
    ret,frame=cap.read() #videoyu okumaya başlıyor, frame o anki kare, ret bir sonraki
    if ret: # ret varsa göster, ret bir sonraki kareye bakıyor yani hala video devam ediyorsa
        cv2.imshow("Demo",frame)
    else:
        break
    key=cv2.waitKey(10)
    if key==ord("q"):
        break
cv2.destroyAllWindows()
```

```python
import cv2
cap = cv2.VideoCapture("cars.mp4")
while True:
    ret,frame=cap.read() #videoyu okumaya başlıyor, frame o anki kare, ret bir sonraki
    if ret: # ret varsa göster, ret bir sonraki kareye bakıyor
        cv2.imshow("Demo",frame)
        gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY) #siyah beyaz çevirdik
        cv2.imshow("SB",gray)
    else:
        break
    key=cv2.waitKey(10)
    if key==ord("q"):
        break
cv2.destroyAllWindows()
```

```python
#Hareket algılama
import cv2
import numpy as np

cap=cv2.VideoCapture(0) #0.kameradaki hareketi anlıyor
#cap=cv2.VideoCapture("cars.mp4") #arabadaki hareketi anlayacak
#cap=cv2.VideoCapture("airplanes.mp4") #uçaktaki hareket
ret1, frame1 = cap.read()
ret2, frame2 = cap.read()

while True:
    frame1_gray = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
    frame2_gray = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
    frame1_blur = cv2.GaussianBlur(frame1_gray, (21,21),0)
    frame2_blur = cv2.GaussianBlur(frame2_gray, (21,21),0)
    diff=cv2.absdiff(frame1_blur, frame2_blur)
    cv2.imshow("Motion", diff)
    frame1 = frame2
    ret,frame2=cap.read()
    if not ret:
        cap.release()
        break
    key=cv2.waitKey(10)
    if key==ord("q"):
        cap.release()
        break
cv2.destroyAllWindows()
```
