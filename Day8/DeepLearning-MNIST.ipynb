```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist
```

```python
(train_images,train_labels),(test_images,test_labels)=mnist.load_data()
```

```python
#Normalize data
```

```python
train_images=train_images/255.0 # RGB için 256 değer alabiliyor. 0-255 arasında oluyor. ben 255'e böldüğümde 1 oluyor. yani  0-1 arasına almış oluyorum veriyi
test_images=test_images/255.0
```

#### Convolutional Neural Networks CNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, InputLayer,Reshape,MaxPooling2D,Flatten
```

```python
model=Sequential()
model.add(InputLayer(input_shape=(28,28))) #shape 'de 28,28 yazıyordu ordan geliyor
model.add(Reshape(target_shape=(28,28,1)))
model.add(Conv2D(filters=12,kernel_size=(3,3),activation='relu')) #kernel_size 3*3 filtre slayttaki
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Flatten())
model.add(Dense(10))
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),optimizer="adam",metrics="accuracy")
```

```python
model.fit(train_images,train_labels,validation_split=0.10,epochs=10)
```
