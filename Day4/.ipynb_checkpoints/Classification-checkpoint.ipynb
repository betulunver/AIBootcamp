##### Feature Engineering

```python
df['Attrition'] 
```

```python
## Attrition sütununa karşılık gelen hücrelerde "yes" ve "no" yazmakta. bunları rakama çevirmemiz lazım. işten ayrılıyorsa yani "yes" ise 1, "no" ise 0 yapalım
```

```python
# dictionary : sözlük tanımlıyoruz
d={"Yes":1,'No':0}
```

```python
 df['Attrition']=df['Attrition'].map(d) #daha sonra d 'yi map ile (for döngüsü gibi) eşitliyor. yes bulduğu yere 1 no bulduğu yere 0 
```

```python
df['Attrition']
```

```python
#df["Attrition"]=df["Attrition"].astype(int64) ## float64 olan ver, tipini int'e çevirecektik ama henüz 265 tane boş alan olduğu için hata verdi. doldurduktan sonra yapıcaz
```

```python
empty=df[df['Attrition'].isnull()] #boş olanları al
```

```python
full=df[df['Attrition'].notnull()] # bol olmayanları al
```

```python
#ben tüm kolonları x'e eşitlemek istiyorum (Attrition hariç)
x=full.drop('Attrition',axis=1) # axis 0 dersek satir  , 1 dersek sütun siliyor
y=full['Attrition']
```

```python
x.sample(5)
```

```python
##tabloda gender, jobRole gibi string ifadeler var bunları hep int'e çevirmemiz lazım
```

```python
x['BusinessTravel'].value_counts()
```

```python
x.shape
```

```python
x=pd.get_dummies(x,drop_first=True)
```

```python
x.head()
```

```python
x.shape
```

```python
#her yazı değeri için sütun ekledi ve sütun sayısı 46'ya çıktı. 
#Sparse Matrix(hiçbirinin bir diğerine üstünlüğü olmayan (kategoriler) string ifadeleri sayısal veri olarak ele alabilmek için )
#n-1 sütun oldu,çünkü hepsinden 0 aldıysa son seçenek 1'dir. bunu ototmatik anladığı için o sütunu yazmadı örn: Non-Travel
```

##### Classification

```python
#Classification Algoritmalarını ekliyoruz
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
g=GaussianNB()
b=BernoulliNB()
```

```python
g.fit(x,y)
b.fit(x,y)
```

```python
# Gaussian -> 1+2+3+4+5+......+100= 
#100+99+98+....+1=
#n(n+1)/2 formülünü buluyor
#Bernoulli -> (aynı aileden)25 tane matematikçi yetiştirmiş
#bernoulli prensibine göre uçaklar uçuyor,uçaklar nasıl uçar, uçakların kanatların üzerindeki basıcı aşağıdaki basıncınndan az olursa uçuyor
```

```python
tahmin1=g.predict(x)
tahmin2=b.predict(x)
```

```python
#Classification -> Accuracy, Precision, Recall, F1 score ve Confusion Matrix
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
accuracy_score(tahmin1,y)
```

```python
confusion_matrix(tahmin1,y)
```

```python
#soldan çapraz doğru tahminler (848+126), sağdan çapraz yanlış tahminler (59+172)
```

```python
print(classification_report(tahmin1,y))
```

```python
sns.heatmap(confusion_matrix(tahmin1,y),annot=True)
```

```python
accuracy_score(tahmin2,y)
```

```python
confusion_matrix(tahmin2,y)
```

```python
sns.heatmap(confusion_matrix(tahmin2,y),annot=True)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris') ##kayıtlı olan data setleirni yükleyebiliyoruz
```

```python
df.head()
```

```python
sns.countplot(df["species"])
```

```python
sns.pairplot(df,hue="species")
```

```python
from sklearn.svm import SVC
```

```python
s=SVC()
```

```python
x=df.drop("species",axis=1)
y=df['species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
pred=s.fit(x_train,y_train).predict(x_test) #tek seferde hem fit hem predict yaptık
```

```python
confusion_matrix(y_test,pred)
```
