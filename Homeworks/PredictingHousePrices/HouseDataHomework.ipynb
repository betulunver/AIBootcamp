### Feature Enginerring

```python
df.isnull().sum()
```

```python
df["zipcode"]=df["zipcode"].astype("object") # zipcode int veri tipinde ama bir üstünlük sayılmadığı için stringe çeviriyoruz
```

```python
df["age"]=2022-df["yr_built"]
```

```python
df["bedrooms"]=df["bedrooms"]**2
df["bathrooms"]=df["bathrooms"]**2
df["sqft_living"]=df["sqft_living"]**2
df["floors"]=df["floors"]**2
df["view"]=df["view"]**2
```

```python
#sns.pairplot(df);
#burada bütün sütunların diğer sütunlara göre değişimini gösterir.
```

## Modelling

```python
#Korelasyon çeşitleri = Pearson, Spearman, Kendall
```

```python
## !pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
#Seed
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x=df[["sqft_living","grade","sqft_above","bathrooms","view",
      "bedrooms","waterfront","floors","zipcode","age","basement","renovated","condition"]] 
#tek parantez koyarsak diziye/ seriye dönüşüyor. tablo şeklinde göstermiyor


y=df[["price"]]
```

```python
x=pd.get_dummies(x,drop_first=True) #
```

```python
df[df["price"]>3000000]
```

```python
df[df["price"]<300000]
```

```python
df[df["bathrooms"]<1]
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=lr.fit(x_train,y_train)
```

```python
tahmin=model.predict(x_test)
```

```python
(mean_squared_error(tahmin,y_test))**0.5
```

```python
r2_score(tahmin,y_test)
```

```python
from sklearn.linear_model import Ridge, Lasso
#tek bir kaynağa güvenmedik ve Ridge ve lasso algoritmalarıyla da denedik.
#Ridge karesini alıyor
#Lasso mutlak değerini alıyor
```

```python
R=Ridge()
L=Lasso()
```

```python
R.fit(x_train,y_train)
#modeli eğittik
```

```python
tahmin2=R.predict(x_test)
#tahmin ettirdik.
```

```python
(mean_squared_error(tahmin2,y_test))**0.5
```

```python
r2_score(tahmin2,y_test)
```

```python
L.fit(x_train,y_train)
```

```python
tahmin3=L.predict(x_test)
```

```python
(mean_squared_error(tahmin3,y_test))**0.5
```

```python
r2_score(tahmin3,y_test)
```
